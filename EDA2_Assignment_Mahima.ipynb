## Insight : This will show the count of each unique value.(Smoking , Obesity and Diabetes)

## Insight : This will show the percentage of each unique value.(Smoking , Obesity and Diabetes)

## Bar plot :

```python
risk_factors = ['Smoking', 'Obesity', 'Diabetes']

# Count the "Yes"/"No" values for each
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i, factor in enumerate(risk_factors):
    df[factor].value_counts().plot(kind='bar', ax=axs[i], color=['red', 'blue'])
    axs[i].set_title(f'{factor} Distribution')
    axs[i].set_ylabel('Count')
    axs[i].set_xlabel(factor)
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=0)

plt.tight_layout()
plt.show()
```

## Insight : These bar charts will visually depict the frequency of each category (its show the count of No is higher as compaired to Yes in smoking , obesity and diabetes)

## TSH_Level	, T3_Level , T4_Level and Nodule_Size :

## Box Plot

```python
sns.boxplot(data = df [['TSH_Level', 'T3_Level', 'T4_Level', 'Nodule_Size']])
plt.title('Age Distribution by Thyroid Cancer Risk')
plt.xlabel('Risk Category')
plt.ylabel('Age')
plt.show()
```

## Insight : Display the risk category of t4 level is higher as compaired to the other level.

# Bar plot 

```python
#Group by Diagnosis and plot average hormone levels
hormone_means = df.groupby('Diagnosis')[['TSH_Level', 'T3_Level', 'T4_Level']].mean()

hormone_means.plot(kind='bar', figsize=(10, 6))
plt.title('Average Hormone Levels by Diagnosis')
plt.ylabel('Hormone Level')
plt.xlabel('Diagnosis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Insight :  show the t4 level is higher as compaired to t3 level and TSH level base on diagnosis.

## Pie Plot

```python
thyroid_cancer_risk_counts = df['Thyroid_Cancer_Risk'].value_counts()

plt.pie(
    thyroid_cancer_risk_counts,
    labels = thyroid_cancer_risk_counts.index,
    autopct ='%1.1f%%',       
    startangle = 140,          
    colors = ['green', 'orange','blue'],  
    wedgeprops = {'edgecolor': 'black'} 
)
plt.title('Thyroid Cancer Risk Distribution')
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()
```

## Insight : to show the count percentage of Thyroid_Cancer_Risk(The low count is higher as compaired to other)

# Box plot 

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Family_History', y='Thyroid_Cancer_Risk', data=df)
plt.title('Family History vs Cancer Risk')
plt.tight_layout()
plt.show()
```
